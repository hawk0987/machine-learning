```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 유방암 데이터셋 로드
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

# train, test 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=26)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

tree4 = DecisionTreeClassifier(max_depth=4)
tree4.fit(X_train, y_train)
print('학습 점수 :', tree.score(X_train, y_train))
print('테스트 점수 :', tree.score(X_test, y_test))

print()
print('학습 점수 :', tree4.score(X_train, y_train))
print('테스트 점수 :', tree4.score(X_test, y_test))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,9))

plot_tree(tree4, class_names=['neg', 'pos'], feature_names=cancer.feature_names, impurity=False, filled=True, rounded=True, fontsize=10)
plt.show()
```

```python
tree4.feature_importances_
```

```python
def plot_feature_importance(model):
  n_features = cancer.data.shape[1]
  plt.barh(np.arange(n_features), model.feature_importances_, align='center')
  plt.yticks(np.arange(n_features), cancer.feature_names)
  plt.xlabel('importance')
  plt.ylabel('feature')
  plt.ylim(-1, n_features)

plot_feature_importance(tree4)
```
